```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def country_visual(country):
    dataset = pd.read_csv('/Users/hilmi/Dev/Omdena_WOE/{}_flood.csv'.format(country),parse_dates=['created_at'])
    grp = dataset.groupby(pd.Grouper(freq='1H', key='created_at'))
    plt.figure(figsize=(12,6))
    plt.plot(grp.count()['id'])
    plt.ylabel('Tweets per hour')
    plt.title("{} Flood Tweets Per Hour".format(country))
```

```python
countries_list = pd.read_excel("Countries.xlsx")
countries_list =[country for country in countries_list["Countries"]]
countries_list
```

```python
from os import listdir
from os.path import isfile, join
```

```python
filelist = [f for f in listdir('/Users/hilmi/Dev/Omdena_WOE') if isfile(join('/Users/hilmi/Dev/Omdena_WOE',f))]
for country in countries_list:
    if '{}_flood.csv'.format(country) in filelist:
        country_visual(country)
```
