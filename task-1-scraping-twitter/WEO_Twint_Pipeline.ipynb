```python
#Cities list 
Countries_list =[country for country in countries_list["Countries"]]
Countries_list
```

```python
def scrape_by_country(keyword, since, country):
    print(country)
    c = twint.Config()
    
    c.Lang = "en"
    c.Search = keyword 
    c.Near = country
    c.Since = since
    c.Hide_output = True
    c.Store_csv = True
    c.Output = "{}_{}.csv".format(country, keyword)
    print("{}_{}.csv".format(country, keyword))
    
    twint.run.Search(c)
```

```python
keyword = "flood"
since = "2021-01-01"
for country in Countries_list:
    scrape_by_country(keyword, since, country)
```
