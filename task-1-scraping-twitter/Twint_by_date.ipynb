```python
import twint
import nest_asyncio
import pandas as pd

nest_asyncio.apply()
```

```python
# Extract historical tweets by hashtag and date
def twint_extract(hashtag, from_date, to_date):
    c = twint.Config()
    c.Search = '#' + hashtag
    
    c.Since = from_date
    c.Until = to_date
    
    c.Store_csv = True
    c.Output = hashtag + '.csv'
    
    twint.run.Search(c)
```

```python
# Call function
twint_extract('flood', '2019-06-01', '2019-07-01')
```
