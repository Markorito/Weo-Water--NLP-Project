```python

for subdir, dirs, files in os.walk(dir):
    print(f'subidr: {subdir}')
    hashtag = subdir.split("/")[-1]
    print(f"hashtag: {hashtag}")
    final_dir = r'/content/gdrive/My Drive/Omdena/task-5-scraping-facebook/instagram_scraper/output_cleaned/' + hashtag
    for file in files:
        print(f"file: {file}")
        try:
            if os.path.exists(f'{final_dir}\\{file}'):
                continue
            df = pd.read_csv(os.path.join(subdir, file))
            df_len = len(df)
            # add a column to be updated if an element has been skipped
            df['skipped'] = False
            for count, (id, permalink) in enumerate(zip(df['id'], df['permalink'])):
                # print(permalink)
                # webbrowser.get('edge').open_new(permalink)
                display(Javascript('window.open("{url}");'.format(url=permalink)))
                print()
                is_useful = input("is the post useful? [y/n/s (skip)]")
                # print(is_useful)
                is_useful = is_useful.lower().strip()
                if is_useful == 'n' or is_useful == 'no':
                    print("drop")
                    df = df.drop(df[df['id'] == id].index)
                elif is_useful == 's' or is_useful == 'skip':
                    print('skip')
                    df.loc[df['id'] == id, 'skipped'] = True
                elif is_useful != 'y' and is_useful != 'yes':
                    raise('wrong letter')
                print(f'remaining {df_len - count}')
            if not os.path.isdir(final_dir):
                os.mkdir(final_dir)
            df.to_csv(f'{final_dir}/{file}', index=False, header=True)

        except Exception as e:
            # If the file is empty catch error
            print(f"exc= {e}")
```
