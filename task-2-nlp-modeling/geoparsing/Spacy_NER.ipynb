```python
text = df['article_content'].iloc[3]
import pandas as pd
doc = nlp(text)

entities = []
labels = []
position_start = []
position_end = []

for ent in doc.ents:
    entities.append(ent)
    labels.append(ent.label_)
    position_start.append(ent.start_char)
    position_end.append(ent.end_char)
    
d = pd.DataFrame({"Entities" : entities, "Labels" : labels, "Position_Start" : position_start, "Position_End" : position_end})

d
```
