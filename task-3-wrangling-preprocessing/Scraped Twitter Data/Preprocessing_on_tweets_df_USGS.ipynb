```python
import pandas as pd
import numpy as np
import csv
from tqdm import tqdm
import base64
from github import Github
from io import BytesIO
import requests
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,RegexpTokenizer
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Username of your GitHub account
 
username = ''
 
# Personal Access Token (PAO) from your GitHub account
 
token = ''
 
# Creates a re-usable session object with your creds in-built
 
github_session = requests.Session()
github_session.auth = (username, token)
```

```python
def download_data (url):

    url = url # Make sure the url is the raw version of the file on GitHub
    download = github_session.get(url).content

    df= pd.read_csv(BytesIO(download))
    
    return df
```

```python
url = 'https://raw.githubusercontent.com/OmdenaAI/WeoWater/main/task-1-scraping-twitter/tweets_df_USGS_line1.csv'
tweets_df_USGS= download_data(url)
```

```python
#droping unwanted columns
tweets_df_USGS.drop(['Unnamed: 0','Tweet Coordinates'],axis=1)
```

```python
#checking null values
tweets_df_USGS.isnull()
```

```python
#Fill null values in Place column
tweets_df_USGS['Place Info'].fillna('Not Specified',inplace=True)
```

```python
#Clean the tweet text using nltk corpus and stopwords
clean_tweets=[]
for tweets in tweets_df_USGS['Tweet Text']:
  tokenizer= RegexpTokenizer(r'\w+')
  token_words= tokenizer.tokenize(tweets)
  token_with_stopword= [word for word in token_words if not word in stopwords.words()]
  clean_tweets.append(token_with_stopword)
```

```python
#Add new column of cleaned tweets
tweets_df_USGS['Cleaned Tweets']= clean_tweets
```

```python
tweets_df_USGS.drop('Unnamed: 0',axis=1)
```

```python
#tagged the data with new label- flood,drought, None
category=[]
for data in tweets_df_USGS['Cleaned Tweets']:
  if any(x in data for x in ['Flood','flooding','flooded','flood']):
    category.append('flood')
  elif 'drought' in data:
    category.append('drought')
  else:
    category.append('None')
```

```python
#Added new column category- which have new tagged label
tweets_df_USGS['category']= category
```

```python
tweets_df_USGS['category'].unique()
```

```python
tweets_df_USGS.to_csv('Preprocessed_tweets_df_USGS.csv',index=False)
```
