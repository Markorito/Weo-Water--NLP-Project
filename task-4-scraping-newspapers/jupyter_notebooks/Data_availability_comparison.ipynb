```python
# countries covered in google news in English
gn_en_list = pd.read_csv('./Countries and languages supported by gn/gn_english.txt',sep='\t')
gn_en_list.fillna(value='English',inplace=True)
gn_en_list.shape
```

```python
#countries covered in google news in other languages (if covered in English and other language, only English is taken)
gn_other_list = pd.read_csv('./Countries and languages supported by gn/gn_other_languages.txt',sep='\t')
gn_other_list.shape
```

```python
gn_countries = gn_en_list.append(gn_other_list,ignore_index=True)
gn_countries['News_source'] = 'Google News'
gn_countries.shape
```

```python
all_countries = pd.read_csv('./Countries and languages supported by gn/all_countries.txt',sep='\t',dtype={'Country':'str','Language':'str','News_source':'str'})
all_countries.shape
```

```python
#The names of following countries are different/not present in all countries list
gn_countries[~gn_countries['Country'].isin(all_countries['Country'])]
```

```python
# joining both the dataframes
all_countries = pd.concat([gn_countries,all_countries])
all_countries.drop_duplicates(subset='Country',inplace=True,keep='first')
all_countries.shape
```

```python
all_countries
```

```python
# WRI (World Resources Institute) list of 15 countries most affected by floods. 
# It accounts for 80% of population exposed to river flood risk
# https://www.wri.org/insights/worlds-15-countries-most-people-exposed-river-floods
wri_flood_countries = ['India','Bangladesh','China','Vietnam','Pakistan','Indonesia','Egypt','Myanmar','Afghanistan','Nigeria','Brazil','Thailand','Congo (Republic of the)','Iraq','Cambodia']
```

```python
all_countries['WRI_flood_list'] = all_countries['Country'].apply(lambda x:x in wri_flood_countries)
```

```python
# List of countries covered by google news
all_countries[(all_countries['News_source']=='Google News') & (all_countries['Language']!='English')]
```

```python
# List of countries in WRI flood list
all_countries[all_countries['WRI_flood_list']==True]
```

```python
# World Atlas list of 10 most drought prone countries of the world 
# https://www.worldatlas.com/articles/the-most-drought-prone-countries-in-the-world.html
wa_drought_countries = ['Ethiopia','Sudan','Eritrea','Afghanistan','China','Pakistan','Iran','Somalia','Uganda','Morocco']
```

```python
all_countries['WA_drought_list'] = all_countries['Country'].apply(lambda x:x in wa_drought_countries)
```

```python
# List of countries in World Atlas list
all_countries[all_countries['WA_drought_list']==True]
```

```python
all_countries['historical_flood_desinventar'] = all_countries['Country'].apply(lambda x:x in historical_flood_desinventar)
all_countries
```

```python
all_countries['historical_drought_desinventar'] = all_countries['Country'].apply(lambda x:x in historical_drought_desinventar)
all_countries
```
